```python
df['method']=df['method'].fillna('')
df['city']=df['city'].fillna('')
df['result_margin']=df['result_margin'].fillna('0')
df['player_of_match']=df['player_of_match'].fillna('')
df['winner']=df['winner'].fillna('')
df['target_overs']=df['target_overs'].fillna('0')
df['super_over']=df['super_over'].fillna('')
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df3=df1.merge(df, left_on='match_id',right_on='id',how='inner')
```

```python
df3.head(1)
```

```python
df3.columns
```

```python
team_runs=df3.groupby('batting_team')['total_runs'].sum().reset_index()
team_runs=team_runs.sort_values(by='total_runs',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(data=team_runs, x='batting_team',y='total_runs',palette='bright',hue='batting_team')
plt.xticks(rotation=45,ha='right')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Total Runs By IPL team")
plt.show()
```

```python
wining_team=df3['winner'].value_counts().head(10)
sns.barplot(x=wining_team.values, y=wining_team.index,palette='bright')
plt.title("Most Successful Team")
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(data=df3,x='toss_decision',hue='match_type')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Toss Decision by match type")
plt.suptitle("countplot")
plt.show()
```

```python
df3['toss_win_match_win'] = df3['toss_winner'] == df3['winner']
toss_win_ratio = df3['toss_win_match_win'].value_counts(normalize=True)*100

ax=sns.barplot(x=toss_win_ratio.index, y=toss_win_ratio.values)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Did Toss Winner Win the Match?")
plt.ylabel("Percentage")
plt.show()
```

```python
plt.figure(figsize=(8,5))
bastman_runs=df3.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=bastman_runs.values, y=bastman_runs.index,palette='bright')
plt.title("Top 10 Run scorers")
plt.xlabel("Total runs")
plt.show()
```

```python

wickets=df3[df3['is_wicket']==1].groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
ax=sns.barplot(x=wickets.values, y=wickets.index,palette='bright')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Top 10 Wicket Takers")
plt.xlabel("Wicket takers")
plt.show()
```

```python
top_bowlers = df3.groupby('bowler')['is_wicket'].sum().reset_index().sort_values(by='is_wicket', ascending=False).head(10)
sns.barplot(data=top_bowlers, x='bowler', y='is_wicket')
plt.xticks(rotation=45)
plt.title("Top 10 Bowlers by Total Wickets")
plt.suptitle("bar chart")
plt.show()
```

```python
plt.figure(figsize=(12,5))
season_runs=df3.groupby('season')['total_runs'].sum().reset_index()
sns.lineplot(data=season_runs, x='season',y='total_runs',marker ='o')
plt.title("Total Runs per Season")
plt.suptitle("bar chart")
plt.show()
```
